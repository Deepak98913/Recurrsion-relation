# 1.  Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.
 
Solution :-   

To find the value of T(2) for the given recurrence relation \( T(n) = 3T(n-1) + 12n \), we can use the initial condition \( T(0) = 5 \).

Given \( T(n) = 3T(n-1) + 12n \), let's first find \( T(1) \) using the recurrence relation:

\[ T(1) = 3T(1-1) + 12(1) = 3T(0) + 12 = 3(5) + 12 = 15 + 12 = 27 \]

Now, to find \( T(2) \):

\[ T(2) = 3T(2-1) + 12(2) = 3T(1) + 24 = 3(27) + 24 = 81 + 24 = 105 \]

So, the value of \( T(2) \) for the given recurrence relation is \( \boxed{105} \).

To solve the recurrence relation \( T(n) = 2T(n/2) + n \) using the substitution method, we'll follow these steps:

1. Perform substitution repeatedly until we reach the base case.
2. Use the base case to work backward and simplify the expression.

Given the recurrence relation \( T(n) = 2T(n/2) + n \), let's start by substituting \( n/2 \) for \( n \) until we reach the base case:

\[
\begin{align*}
T(n) &= 2T(n/2) + n \\
&= 2[2T(n/4) + n/2] + n \\
&= 2^2T(n/2^2) + 2n + n \\
&= 2^2T(n/2^2) + 3n \\
&= 2^2[2T(n/2^3) + n/2^2] + 3n \\
&= 2^3T(n/2^3) + 3n + \frac{3n}{2} \\
&\dots \\
&= 2^kT(n/2^k) + kn
\end{align*}
\]

We've reached \( T(n/2^k) \), which is typically the base case. Now, let's determine the value of \( k \) when \( n/2^k = 1 \):

\[
\frac{n}{2^k} = 1 \implies n = 2^k \implies k = \log_2(n)
\]

Using the base case \( T(1) \), we have:

\[
T(1) = 2^kT(n/2^k) + kn = 2^{\log_2(n)}T(1) + n \log_2(n) = nT(1) + n \log_2(n)
\]

Given the base case \( T(1) \), typically \( T(1) = \text{constant} \), we can say \( T(1) = c \). Therefore:

\[
T(n) = nT(1) + n \log_2(n) = cn + n \log_2(n)
\]

So, the solution to the recurrence relation \( T(n) = 2T(n/2) + n \) using the substitution method is cn + nlog_2(n)
